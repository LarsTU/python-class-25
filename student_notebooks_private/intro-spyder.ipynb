### Getting started

#### Running Code
+ create a new .py-file 
+ type ``print("Hello World")`` into the editor
+ run the **whole script** with `F5` (shortcut) 
    + We can see in the Python Console
        + the output, 
        + the path of the file that we are running and
        + the working directory where this code was run.
+ run code of a **single line** (instead of the whole script):
    + go into the respective line and press `F9` (shortcut) 
+ run code of a **cell** in the script:
  + go to the editor and type below `print("Hello World")` the lines as indicated in the following code block
  + go into the line with `#%% first cell`
  + run a cell with `Ctrl`+`Enter` (shortcut)

```python
#%% first cell
print("This is cell 1")

#%% second cell
print("This is cell 2")

#%%
```

+ run code **interactively in the IPython Console**
    + type "Hello world" in the Console and hit enter
+ A summary of all shortcuts can by found in `Help > Shortcuts Summary`

#### Defining variables

+ you can define variables from both, the editor or the console:
    + type `a = 10` below the last `#%%` in your file and run the line (`F9`)
    + type `b = 20` in the IPython console and hit enter
+ have a look into the `Variable Explorer` (upper right pane)
    + for both variables `a` and `b` you see the name, type, size and value 
+ Next, define a list in the editor below `a = 10` by typing `c = [1, 2, 3]` and run the line (`F9`)  
+ You can see in the `Variable Explorer` that the type of the variable is list and the size is 3.

#### Changing and removing variable values

+ We can **change the values** of the variables in the Variable Explorer too
    + double-click on `a`, change the value to `30` and hit enter
    + now type the variable in the IPyhton Console and check that the value of `a` has indeed changed
    + changing more complex types, e.g. a list, will open a viewer,
        + double-click on the variable `c` and change the third value to `4`, then type `c` in the IPython Console and check whether it has really the values `c = [1,2,4]`
+ Next, we **remove** the variable `b`
    + variant 1) right-click on the variable `b` in the Variable Explorer and selecting the option Remove
    + variant 2) type in the IPython Console `del b`
+ Check in the IPthon Console whether the variable was actually deleted (i.e., you should get an error message indicating that b is not defined)

#### Getting help

+ import numpy and define a new variable:
    + import numpy in the python script by typing `import numpy as np`
    + define a new variable: `v = np.array([1,2,3])`
    + select and run both lines (use `F9`)
+ you can get `help` for an object in two ways:
    + select the written object in your file and press `Ctrl-I` (or `Cmd-I` on macOS): try it for `np.array`
    + in the help pane, select the tab `Help`, then you will find in the upper part of the pane a Source dropdown option. Change it to from *Editor* to *Console* and type `np.array` in the *Object* box

### Customize Spyder

**Font**
+ go to *Tools > Preferences* and select the *Appearance* entry. In the bottom you will find the section *Fonts*.
+ You can change both the style and the size of the font for both plain and rich text.
+ When you press the `Apply` button in the lower right, you can directly see how the change looks like

**syntax highlighting theme**
+ in the same dialog, you find also the section *syntax highlighting theme*
+ choosing a theme will show you directly in the preview how it looks like
+ select a theme and the click `Apply` to restart Spyder to apply the new theme.

**Select/Rearrange Panes**
+ you can also rearrange the panes in Spyderâ€™s main window
+ Go to *View > Panes*: here you can select which panes you want to see. 
+ By default, the panes and toolbars are locked so they canâ€™t be moved accidentally.
+ In *View > Unlock panes and toolbars* allows you to move them freely anywhere on the window by dragging them from the top and dropping them at any position you like.

**Further customization options for each pane**
+ Go to *Tools > Preferences* and select the *Editor* entry. Here further adjust the editor appearance (the same is possible for the other panes of course)

**Back to the roots**
+ if you ever want to return to its default configuration, you can always reset the window layout under *View > Window Layouts* or your entire Spyder configuration in *Tools > Preferences* with the `Reset to Defaults` button in the lower left corner.

### Our first modules
+ create a new folder called `mypackage_temp`
+ open a new Python script which we will save as `simulations.py` in the new folder 
+ include the following code in this script ([code source](https://github.com/anishsatalkar/Galton-Board-Python/tree/master))

```python
#%% simulations.py

# import necessary libraries
from random import randint


def galton(num_bins, num_balls):
    # create the number of bins
    bins = [0]*(num_bins+1)
    # for each ball
    for _ in range(num_balls):
        pos = 0
        # go over each bin
        for _ in range(num_bins):
            coin_flip = randint(0,1)
            # if yes, increase bin-counter by 1
            if coin_flip == 1:
                pos += 1
        # save final result
        bins[pos] = bins[pos] + 1
    
    return bins
```

+ create a further script, called `plotting.py`, and save it as well in the folder `mypackage_temp` 
+ include the following code:

```python
#%% plotting.py


def print_galton_board(data):
    for val in data:
        # print for each bin a dot for each ball
        for _ in range(0,val):
            print(".", end="")
        # show total number of balls each bin at the end of the line
        print(val)
```
