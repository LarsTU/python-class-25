### Python environment & package management: Anaconda, conda, & pip
**Anaconda**
+ environment and package manager: comes already with a lot of tools that we will use (such as Spyder, Jupyterlab, etc.)
+ easy creation of **environments** to organise your projects 
+ however, there are several other package/environment managers that you can use
    +  [`mamba`](https://mamba.readthedocs.io/en/latest/installation/mamba-installation.html)
    +  [`uv`](https://docs.astral.sh/uv/)
    +  [`poetry`](https://python-poetry.org/docs/)
    +  [`pdm`](https://pdm-project.org/en/latest/)
    +  [`pixi`](https://pixi.sh/dev/)
+ Note: Ananconda changed their Terms of Services (see e.g. [this Twitter-Thread](https://www.theregister.com/2024/08/08/anaconda_puts_the_squeeze_on/))
    + main issue: non profit organisations with more that 200 employees need to buy commercial licenses (result: most institutes will ban anaconda and use alternatives)
    + lessons learned: don't make yourself dependent on one specific tools and make sure that the tools that you use are driven by community work/open science
      
**Installing packages via `conda` and `pip`**
+ when we want to install a module via Anaconda's package manager we use `conda`
+ `pip` can install anything from the [Python Package Index (`PyPI`)](https://pypi.org/) which is a repository of software for the Python programming language (equivalent to CRAN in R)
+ `pip` vs. `conda`
    + pip installs Python packages
    + conda installs packages which may contain software written in any language. (For example, before using pip, a Python interpreter must be installed via a system package manager or by downloading and running an installer. Conda on the other hand can install Python packages as well as the Python interpreter directly.)
    + Occasionally a package is needed which is not available as a conda package but is available on PyPI and can be installed with pip. In these cases, it makes sense to try to use both conda and pip.

### Git, Git Bash, & GitHub
+ [`Git`](https://git-scm.com/) for version control 
    + Version control is a system that records changes to a file or set of files over time so that you can recall specific versions later
    + Git is a system that supports you in tracking and tracing the history of your files
+ to run Git in Windows from the command line we need [`Git Bash`](https://gitforwindows.org/)
    + GitBash is an application for Microsoft Windows environments which provides an emulation layer for a Git command line experience.
    + We can install it directly, but it comes also directly when we use GitHub Desktop
+ [`GitHub`](https://github.com/) for collaborative work; sharing your work with the community & backup

    + GitHub Desktop is a Git client for streamlining and enhancing your local code management process
