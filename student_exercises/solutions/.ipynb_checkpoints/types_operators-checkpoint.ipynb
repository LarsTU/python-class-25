```python
# exercise 1
# Answer: If two multiples are equally close, rounding is done toward the even choice
print( round(42.5) )
print( round(43.5) )

# exercise 2
print( +2 )
print( -2 )
print( 2++2 )
print( 2--2 )

# exercise 3
print( type(765) )
print( type(2.718) )
print( type('2 pi') )
print( type(abs(-7)) )
print( type(abs(-7.0)) )
print( type(abs) )
print( type(int) )
print( type(type) )

# exercise 4
## 1) How many seconds are there in 42 minutes 42 seconds?
secs = 42 * 60 + 42
print( secs )
## 2) How many miles are there in 10 kilometers? Hint: there are 1.61 kilometers in a mile.
miles = (1/1.61) * 10 
print( miles )
## 3) If you run a 10 kilometer race in 42 minutes 42 seconds, what is your average pace in seconds per mile?
secs_miles = secs/miles
print( secs_miles )
## 4) What is your average pace in minutes and seconds per mile?
## minutes
print( secs_miles // 60 )
## seconds
print( secs_miles - (6*60) )
print( secs_miles % 60 ) # remainder
## built-in function
## minutes, seconds = divmod(seconds, 60)
print( divmod(secs_miles, 60) )
help( divmod )

# Exercise 5
## %Git Bash
## pip install matplotlib
import matplotlib
print(f'Matplotlib version: {matplotlib.__version__}')

# Exercise 6
# Step 1: Import the pyplot module with the alias plt
import matplotlib.pyplot as plt
# Step 2: Import the numpy library to generate data for the sine function
import numpy as np
# Step 3: Generate data for the sine function
x = np.linspace(0, 2 * np.pi, 100)  # 100 values from 0 to 2pi
y = np.sin(x)  # Sine of each value in x
# Step 4: Create the plot
plt.plot(x, y)
# Step 5: Add labels and a title
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Function')
# Step 6: Display the plot
plt.show()

# Exercise 7:
import numpy as np
# Create a 3x3 matrix of random integers between 1 and 10
matrix = np.random.randint(1, 11, size=(3, 3))
print("Random matrix: \n", matrix)
# Compute and print determinant
determinant = np.linalg.det(matrix)
print(f"Determinant:\n {determinant:.2f}")
```
