### 6. Dog class
+ create a class `Dog` with attributes: name, year_of_birth (optional argument)

````python
# 1) create a class with instance attributes
class Vehicle:
    def __init__(self, max_speed, mileage):
        self.max_speed = max_speed
        self.mileage = mileage

volvo = Vehicle(max_speed=180, mileage=12)

print( dir(volvo) )
print( volvo.max_speed )
print( volvo.mileage )

# 2) inheritance
class Bus(Vehicle):
    # assign default value to capacity
    def seating_capacity(self, capacity):
        return print("Volvo bus with max_speed="+str(self.max_speed)+", mileage="+str(self.mileage),"and capacity="+str(capacity))

volvo_bus = Bus(max_speed=180, mileage=12)
volvo_bus.seating_capacity(capacity=50)

# 3) class variable
class WhiteBus(Bus):
    color = "white"

white_bus = WhiteBus(max_speed=180, mileage=12)
white_bus.color

# 4) Inheritance (2)
class Vehicle:
    def __init__(self, name, mileage, capacity):
        self.name = name
        self.mileage = mileage
        self.capacity = capacity

    def fare(self):
        return self.capacity * 100

class Bus(Vehicle):
    def fare(self):
        amount = super().fare()
        amount += amount * 10 / 100
        return amount

school_bus = Bus("School Volvo", 12, 50)
print("Total Bus fare is:", school_bus.fare())
# type of school_bus
print(type(school_bus))
print(isinstance(school_bus, Vehicle))

# 5) spot mistakes
class Pet:
    def __init__(self, name):  # <<<<<<<<<<<< 
        self.name = name
        
    def move(self):            # <<<<<<<<<<<<
        return f"{self.name} is moving..."

pet = Pet("Bebo")
pet.move()

# 6) a dog class

class Dog:
    def __init__(self, name, year_of_birth=None):
        self.name = str(name)
        self.year_of_birth = int(year_of_birth)

    def get_name(self):
        print("My dog's name is "+self.name.title())

    def get_age(self):
        from datetime import date

        todays_date = date.today()
        age = todays_date.year-self.year_of_birth
        print("My dog is "+str(age)+" years old. In a dog's life this is "+str(age*7)+" years.")

my_dog = Dog("linus", 2017)
my_dog.get_name()
my_dog.get_age()
````
